### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = "|")
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation').age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
u = users.groupby('occupation')
for i,j in u:
    print(i)
    print(j)

#make a new column for conversion:
def convert(c):
    if c == 'M':
        return 1
    if c == 'F':
        return 0
users['gender_num'] = users['gender'].apply(convert)
#find the ratio:
ratio = users.groupby('occupation').gender_num.sum()*100 / users.occupation.value_counts()
#order
ratio.sort_values(ascending = False)
```

```python
ratio = users.groupby('occupation').gender_num.sum()*100 / users.occupation.value_counts()
ratio.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').age.agg(['max','min'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).age.mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_occupation = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
#print(gender_occupation)
occupation_count = users.groupby(['occupation']).agg('count')
#print(occupation_count)
occupation_gender = gender_occupation.div(occupation_count, level ='occupation')*100 #element wise division of the two
#printing specific elelment wise
occupation_gender.loc[:,'gender']
```
