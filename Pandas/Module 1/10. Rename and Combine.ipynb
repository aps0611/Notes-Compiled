# Renaming

 lets you change index names and/or column names

```python
data.rename(columns = {'Kms Driven' : 'RideKm' })
```

rename() lets you rename index or column values by specifying a index or column keyword parameter, respectively. It supports a variety of input formats, but usually a Python dictionary is the most convenient.

```python
data1= data.rename(index = { 0: 'first',
                            1 : 'Second'
})
data1
```

Both the row index and the column index can have their own name attribute. The complimentary rename_axis() method may be used to change these names. For example:

```python
data1.rename_axis('ind',axis = 'rows').rename_axis('col',axis = 'columns')
```

# Combining

When performing operations on a dataset, we will sometimes need to combine different DataFrames and/or Series in non-trivial ways. Pandas has three core methods for doing this. In order of increasing complexity, these are concat(), join(), and merge()

# # Concat:

```python
weather1 = DataFrame({
    'city' : ['Pune','Kolkata','Madras'],
    'Temperature' : [32,45,46],
    'humidity' : [80,60,78]
})
weather1
```

```python
weather2 = DataFrame({
    'city' : ['Delhi','Bhopal','Imphal'],
    'Temperature' : [35,40,28],
    'humidity' : [50,65,90]
})
weather2
```

```python
pd.concat([weather1,weather2])
```

```python
pd.concat([weather1,weather2],ignore_index = True)
```

```python
#pass Keys argument- helps associate keys with each dataframe
```

```python
df = pd.concat([weather1,weather2],keys = ['one','two'])
df
```

# MERGE

```python
pd.merge(tempdf,winddf, on='city') 
#it is looking for city and then joining them
```

```python
tempdf1 = DataFrame({
    'city' : ['Madras','Kolkata','Pune','Buldana'],
    'Temperature' : [32,45,46,50]})
```

```python
pd.merge(tempdf1,winddf,on = 'city') 
#as we see the merge does the intersection and does not return the union as buldana is missing
```

```python
pd.merge(tempdf1,winddf,on = 'city',how ='inner') #inner by default 
```

```python
pd.merge(tempdf1,winddf,on = 'city', how ='left')  #left join
```

```python
pd.merge(tempdf1,winddf,on = 'city', how ='right')  #right join
```

```python
pd.merge(tempdf1,winddf,on = 'city', how ='left', indicator = True)  #inidicator set to true will indicate from where did the data come from
```

```python
pd.merge(weather1,weather2,on = 'city',how = 'outer', suffixes = ('_left','_right'), indicator = True )
```
