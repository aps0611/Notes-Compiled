```python
import pandas as pd
from pandas import DataFrame, Series
import numpy as np
data = pd.read_csv('weather.csv')
data
```

```python
data.pivot(index ='day', columns = 'city')
```

```python
data.pivot(index ='day', columns = 'city', values = 'temperature')
```

```python
df = pd.read_csv('weather2.csv')
df
```

```python
df.pivot(index='city', columns ='day')
```

```python
df.pivot_table(index='city', columns ='day')  #gives the average if there are two values with same date
```

```python
df.pivot_table(index='city', columns ='day', aggfunc = 'sum') #since column is day it will give sum of same day
```

```python
df.pivot_table(index='city', columns ='day', aggfunc = 'count')
```

```python
df.pivot_table(index='city', columns ='day', aggfunc = 'mean') #default
```

```python
df.pivot_table(index='city', columns ='day', margins = 'True')
```

```python
df['day']=pd.to_datetime(df['day'])
df.pivot_table(index = pd.Grouper(freq='M',key='day'), columns ='city')
```
