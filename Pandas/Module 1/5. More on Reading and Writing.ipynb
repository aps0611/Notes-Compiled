```python
data = pd.read_csv('cars.csv',nrows = 3) #It will read only 3 rows
data
```

```python
data = pd.read_csv('cars.csv',na_values=['Tata','Delhi'])
data
```

it is important to notice that na_value replaced the values given e.g. Delhi and Tata with NaN!!

Dictionary gives us more specific use case for converting some values to NaN

```python
data = pd.read_csv('cars.csv',header=1,na_values={'Year':[2011,2012],
                                        'City':['Mumbai','Chennai']})
data

#As seen in the result we can convert specific values in particular columns into NaN values which is not possible in case of general list
```

# Writing

```python
data.to_csv('new.csv', index = False, header = 1,columns =['Brand','Year','Kms Driven','City','Mileage'])

pd.read_csv('new.csv')
```

1. index = False will not write the index
2. header = 1 will select header as Brand and not the wrong row
3. colums = [  ] will write only those columns which are specified in the given list

```python
data.to_excel('new.xlsx',sheet_name = 'carsdata',startrow = 1, startcol=1)
```

Two data frames and then how can we add them to single excel file?

# excel_writer:

```python
with pd.ExcelWriter('new.xlsx') as writer:
    data.to_excel(writer,sheet_name = '001')
    data1.to_excel(writer,sheet_name = '002')

pd.read_excel('new.xlsx',sheet_name = '001').head(3)
```

# Converter function:
####Helps in converting messy data into something useful

```python
def convert_city_cell(cell):
    if cell == '':
        return 'Pune'
    return cell

data = pd.read_csv('new.csv',converters = {
    'City': convert_city_cell
})
data
```
